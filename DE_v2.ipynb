```python
#Tabla de eventos normalizada.
Eventos=csv1[['idEvent','Event']]
Eventos
```

```python
#Tabla de departamentos normalizada.
Departaments=csv1[['idDepartamento','Departamento']]
Departaments
```

```python
#Tabla de fuentes de informaci√≥n.
Source=csv1[['idSource','Source']]
Source
```

```python
df_EventosSource= Source.merge(Eventos, on="idEvent")
```

```python
#Merge entre eventos y tabla de fuentes de informaci√≥n
df_EventosSource
```

```python
#Creaci√≥n de tabla que muestre por cada departamento, el n√∫mero de v√≠ctimas que han habido para cada tipo de evento.
tabla_model = pd.pivot_table(df1, values='Victims', index=['idDepartamento'],
                       columns=['Event'], aggfunc="sum").reset_index()
tabla_model.columns
```

```python
tabla_model
```

#En qu√© departamento se generan m√°s eventos

```python
#Se realiza una tabla que muestre los eventos para cada departamento...
df_EventoDepartamento= Departaments.merge(tabla_model, right_on = 'idDepartamento', left_on='idDepartamento',how='inner')
df_EventoDepartamento
```

```python
#Con el insumo de la tabla anterior, se puede saber en qu√© departamento se generan m√°s eventos de deslizamiento
df2 = df_EventoDepartamento[["Departamento","LANDSLIDE"]].groupby("Departamento").sum().astype(int)
df2
```

```python
df2.sort_values(by = 'LANDSLIDE', ascending=False).head(5)
```

Suma de todos los eventos por cada departamento.

```python
#Suma de todos los eventos (Landslide + ... + etc.), por idDepartamento.
#Con este c√≥digo es posible anexar el tipo de evento deseado, y dar√° la suma por id de departamento.
dftotal = tabla_model['ACCIDENT']+tabla_model['ALLUVION']
dftotal
```

```python
#Conteo de todos los eventos en total por cada uno de los departamentos
#Esto significa que por ejemplo, en amazonas ocurrieron 84 eventos desde deslizamiento hasta aluviones.
SumEventos = df1.groupby("Departamento")["Event"].count()
SumEventos
```

```python
#Departamentos m√°s afectados por eventos, top 5
SumEventos.sort_values(ascending = False).head(5)
```

```python
#Los municipios que m√°s eventos sufrieron, por n√∫mero de eventos.
#Se muestra el sin dato, para no reportados en municipio.

SumEventosMun = df1.groupby("Municipio")["Event"].count()
SumEventosMun.sort_values(ascending = False).head(5)
```
